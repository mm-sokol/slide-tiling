```python
import xml.etree.ElementTree as ET
tree = ET.parse(r'dane\22_09617A1_7.2.xml')
root = tree.getroot()
```

.get("attribute_name") - get a value of attribute
.find("./tag_name") - get a child tag
.findall("./tag_name") - get all matching children
.tag - returns tag

```python
print(root)
```

```python
for gggg in root:
    print(gggg.tag)
```

```python
annotations_group = root.find("./Annotations")
```

```python
for gggg in annotations_group:
    print(gggg.get("Name"))
```

```python
annotations_list = annotations_group.findall("Annotation")
annotations_list
```

```python
for annotation in annotations_list:
    if annotation.get("Type") == "Dot":
        print(annotation.get("Name"))
        coordinate_pair = annotation.find("./Coordinates").find("./Coordinate")
        print(f"({coordinate_pair.get('X')}; {coordinate_pair.get('Y')})")
    if annotation.get("Type") == "Rectangle":
        print(annotation.get("Name"))
        coordinate_pairs = annotation.find("./Coordinates").findall("./Coordinate")
        for coordinate_pair in coordinate_pairs:
            print(f"({coordinate_pair.get('X')}; {coordinate_pair.get('Y')})")
```
